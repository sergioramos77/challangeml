## Principios generales

* **Consistencia.** Todos los objetos (básicos o compuestos) comparten una interfaz consistente compuesta por un conjunto limitado de métodos. Esta interfaz se documenta de manera consistente para todos los objetos.

* **Inspección.** Los parámetros del constructor y los valores de los parámetros determinados por los algoritmos de aprendizaje se almacenan y se muestran como atributos públicos.

* **No proliferación de clases.** Los algoritmos de aprendizaje son los únicos objetos que se representan mediante clases personalizadas. Los conjuntos de datos se representan como matrices NumPy o matrices dispersas SciPy. Los nombres y valores de los hiperparámetros se representan como cadenas o números de Python estándar siempre que sea posible. Esto mantiene scikitlearn fácil de usar y fácil de combinar con otras bibliotecas.

* **Composición.** Muchas tareas de aprendizaje automático se pueden expresar como secuencias o combinaciones de transformaciones de datos. Algunos algoritmos de aprendizaje también se ven naturalmente como metaalgoritmos parametrizados con otros algoritmos. Siempre que sea viable, se implementan y componen tales algoritmos a partir de bloques de construcción existentes.

* **Valores predeterminados sensatos.** Cada vez que una operación requiere un parámetro definido por el usuario, la biblioteca define un valor predeterminado apropiado. El valor predeterminado debe hacer que la operación se realice de una manera sensata (proporcionando una solución base para la tarea a mano).

# Supervised Learning: Simple Linear Regression

 * generate_random = np.random.RandomState(0)

El código genera_random = np.random.RandomState(0) crea un generador de números aleatorios con una semilla específica (en este caso, 0). Esto significa que cada vez que se llama a este generador, producirá los mismos números aleatorios en el mismo orden. Esto es útil para realizar pruebas y asegurarse de que los resultados sean consistentes.

* x = 10 * generate_random.rand(100)

Este código genera una matriz de 100 elementos, cada uno con un valor aleatorio multiplicado por 10. Esto significa que el resultado será una matriz de 100 elementos, cada uno con un valor entre 0 y 10.

```python
import numpy as np
generate_random = np.random.RandomState(0)
x = 10 * generate_random.rand(100)
```

```python
x
```

* y = 3 * x + np.random.randn(100)

Este código genera una lista de 100 números aleatorios que se generan a partir de la multiplicación de 3 por un número x, más un número aleatorio generado por la función np.random.randn(). Esta función devuelve una muestra (de tamaño 1) de una distribución normal estándar. Los resultados se almacenan en la variable y.



```python
y = 3 * x + np.random.randn(100)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (15, 10))
plt.scatter(x, y);
```

## Step 1. Choose a class of model

```python
from sklearn.linear_model import LinearRegression
```

## Step 2. Choose model hyperparameters


El parámetro fit_intercept=True indica que el algoritmo de regresión lineal debe ajustar una línea de ajuste a los datos. Esto significa que el algoritmo calculará un valor para la pendiente y el intercepto para la línea de ajuste, lo que le permitirá predecir los resultados para un conjunto de datos dado.



```python
model = LinearRegression(fit_intercept=True)
```

model

El código model.fit es una función de la biblioteca de sklearn, que se utiliza para entrenar un modelo de aprendizaje automático. Esta función toma como parámetros los datos de entrada, el número de iteraciones a realizar y los parámetros del modelo. Esta función entrena el modelo en base a los datos de entrada y los parámetros especificados, y devuelve un objeto con información sobre el rendimiento del modelo durante el entrenamiento.

```python
model
```

## Step 3. Arrage data into features matrix and target array

Este código toma una matriz x y la reforma en una matriz con una sola columna. El parámetro -1 significa que el número de filas se determinará automáticamente en función del número de elementos en la matriz original. La línea X.shape devuelve la forma de la matriz X, es decir, el número de filas y columnas.

```python
X = x.reshape(-1, 1)
X.shape
```

```python
print(type(X))
print(type(x))
```

```python
X
```

## Step 4. Fit model to data



Model.fit(X, y) es una función de la biblioteca de Python Scikit-Learn que se utiliza para ajustar un modelo de aprendizaje automático a un conjunto de datos. Esta función toma dos argumentos: X, que es un conjunto de datos con características o variables predictoras, y y, que es un vector con etiquetas o valores objetivo. El modelo entonces se ajusta a los datos usando algoritmos de aprendizaje automático como regresión lineal, árboles de decisión o redes neuronales para predecir el valor objetivo.



```python
model.fit(X, y)
```

En scikit-learn, el atributo "coef_" de un modelo es un numpy array que contiene los coeficientes o pesos asociados con las características en un modelo de aprendizaje supervisado, como por ejemplo en una Regresión Lineal.

Estos coeficientes representan la importancia relativa de cada característica en la predicción de la variable objetivo. Un coeficiente positivo sugiere que a medida que una característica aumenta, la variable objetivo también aumentará, mientras que un coeficiente negativo sugiere lo contrario.

```python
model.coef_
```

En scikit-learn, el atributo intercept_ de un modelo es un número que representa el término independiente o "intercepto" en un modelo de aprendizaje supervisado, como por ejemplo en una Regresión Lineal.

El intercepto es un término constante que se agrega a la combinación lineal de las características para producir la predicción. Por lo tanto, representa el valor esperado de la variable objetivo cuando todas las características tienen un valor de cero.

Al igual que con los coeficientes, la interpretación del intercepto depende del contexto y del problema en el que se esté utilizando el modelo.

```python
model.intercept_
```

## Step 5. Apply trained model to new data

El código crea un array de números equidistantes entre -1 y 11. La función np.linspace() crea un array de números equidistantes entre dos límites especificados. En este caso, el array se llama x_fit y contiene números equidistantes entre -1 y 11.

```python
x_fit = np.linspace(-1, 11)
```

```python
x_fit 
```


Este código convierte una matriz de una sola dimensión en una matriz de dos dimensiones. Esto se logra mediante el uso del parámetro "reshape" y especificando "-1" como el primer argumento. Esto significa que el número de filas se ajustará automáticamente para que coincida con el número de elementos en la matriz original. El segundo argumento especificado como "1" indica que la matriz resultante tendrá una sola columna.

```python
X_fit = x_fit.reshape(-1,1)
```

```python
X_fit
```

```python
y_fit = model.predict(X_fit)
```

## Visualise

```python
plt.figure(figsize = (10, 8))
plt.scatter(x, y)
plt.plot(x_fit, y_fit);
```
