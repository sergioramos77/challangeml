```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

aleatorio = np.random.RandomState(0)
x = 10 * aleatorio.rand(100)
```

```python
y = 3 * x**2 + np.random.randn(100)
```

```python
plt.figure(figsize = (10, 10))
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression(fit_intercept=True)
```

```python
modelo
```

```python
X = x.reshape(-1, 1)
X.shape
```

```python
modelo.fit(X, y)
```

```python
modelo.coef_
```

```python
modelo.intercept_
```

```python
x_aj = np.linspace(-1, 11)
```

```python
X_aj = x_aj.reshape(-1,1)
```

```python
y_aj = modelo.predict(X_aj)
```

## Visualise

```python
plt.figure(figsize = (10, 10))
plt.scatter(x, y)
plt.plot(x_aj, y_aj);
```
